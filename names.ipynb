```python
#import dependencies
import pandas as pd
import glob
import os
```

```python
path = r'namesbystate'                     
all_files = glob.glob(os.path.join(path, "*.csv")) 

df_from_each_file = (pd.read_csv(f, header=None) for f in all_files)
```

```python
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True, sort=True)
concatenated_df.head(20)


```
